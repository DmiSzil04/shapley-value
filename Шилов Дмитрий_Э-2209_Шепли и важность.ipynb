```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from statsmodels.api import add_constant
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_validate
from sklearn.utils import shuffle
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/katarina74/ml_lessons/main/lesson_2/data/techparams_train.csv")
X = df.drop(['target', 'index'], axis=1)
y = df['target']
```

```python
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.33, random_state=17)
```

```python
reg_model = LinearRegression().fit(train_X, train_y)
```

```python
reg_model.coef_
```

<h2>Индекс Шепли</h2>

<a href="https://ibb.co/MxJYmp6X"><img src="https://i.ibb.co/B5YX9jN1/2025-04-13-142702.png" alt="2025-04-13-142702" border="0"></a>

```python
def shapley(df, model, column):
    values = []
    mean_factor = np.mean(df.iloc[:, column])
    for i in range(df.shape[0]):
        values.append(model.coef_[column] * (df.iloc[i, column] - mean_factor))
    return np.array(values)
```

```python
shapley(test_X, reg_model, 0)
```

```python
indexes = []
for j in range(test_X.shape[1]):
    indexes.append(shapley(test_X, reg_model, j))
    sns.kdeplot(indexes[j])
    plt.title('Индекс Шепли для признака {}'.format(test_X.columns[j]))
    plt.show()

indexes
```

```python
shap_val = sorted(zip(test_X.columns, [round(np.median(x), 4) for x in indexes]), key=lambda x: abs(x[1]))
shap_val
```

```python
print('Наиболее важный признак: {}'.format(shap_val[-1][0]))
```

<h2>Пермутационная важность</h2>

```python
MSE_base = mean_squared_error(test_y, reg_model.predict(test_X))
print(round(MSE_base, 4))
```

```python
vals = []
for i in range(len(train_X.columns)):
    ds = pd.DataFrame(train_X.drop(columns=train_X.columns[i]).reset_index(drop=True))
    ds.insert(i, train_X.columns[i], pd.DataFrame(train_X.iloc[:, i].sample(frac=1).reset_index(drop=True)).iloc[:, 0])
    regres = LinearRegression().fit(ds, train_y)
    dd = pd.DataFrame(test_X.drop(columns=test_X.columns[i]).reset_index(drop=True))
    dd.insert(i, test_X.columns[i], pd.DataFrame(test_X.iloc[:, i].sample(frac=1).reset_index(drop=True)).iloc[:, 0])
    vals.append(round(mean_squared_error(test_y, regres.predict(dd)), 4))
    
factors_perm = sorted(zip(train_X.columns, vals), key=lambda x: x[1])
for j in factors_perm:
    print(j)
print(f'Сильнее всего качество модели ухудшается при перестановке фактора {factors_perm[-1][0]}')
```
